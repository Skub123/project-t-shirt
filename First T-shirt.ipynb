```python
import matplotlib.pyplot as plt
import PIL
from PIL import Image

def show_pillow_image(image):
    plt.figure() 
    plt.imshow(image)
    #plt.ioff()
    plt.show()

def open_image(file):
    return Image.open(file)        
        
brain_images = []
for i in range(6):
    file = 'brain_images/output-brain-{}.png'.format(i)
    brain_image = open_image(file)
    brain_images.append(brain_image)
    show_pillow_image(brain_image)
     
```

```python
udacity_image = open_image('images/Social2-FB.png')
show_pillow_image(udacity_image)
print('udacity_image.size={}'.format(udacity_image.size))
```

```python
offset = 30
udacity_width = 1040
above_image = udacity_image.crop((offset, 230, offset + udacity_width, 610))

width = 1300
height = int(above_image.size[1] / above_image.size[0] * 1300 + 0.5)
resize = (width, height)
print('resize={}'.format(resize))
above_image = above_image.resize(resize, Image.BICUBIC)

show_pillow_image(above_image)
print('above_image.size={}'.format(above_image.size))
```

```python
offset = 70
below_image = udacity_image.crop((offset, 40, offset + udacity_width, 170))

width = 1300
height = int(below_image.size[1] / below_image.size[0] * 1300 + 0.5)
resize = (width, height)
print('resize={}'.format(resize))
below_image = below_image.resize(resize, Image.BICUBIC)

show_pillow_image(below_image)
print('below_image.size={}'.format(below_image.size))
```

```python
screen_width = 400 #150

pytorch_image = open_image('design2/pytorch.jpeg')
show_pillow_image(pytorch_image)
pytorch_image = pytorch_image.crop((0, 200, pytorch_image.size[0], 1000))
height = int(pytorch_image.size[1] / pytorch_image.size[0] * screen_width + 0.5)
resize = (screen_width, height)
print('resize={}'.format(resize))
pytorch_image = pytorch_image.resize(resize, Image.ANTIALIAS)
show_pillow_image(pytorch_image)

openmined_image = open_image('design2/openmined.png')
height = int(openmined_image.size[1] / openmined_image.size[0] * screen_width + 0.5)
resize = (screen_width, height)
print('resize={}'.format(resize))
openmined_image = openmined_image.resize(resize, Image.ANTIALIAS)
show_pillow_image(openmined_image)

logo3_image = openmined_image.crop((0, 0, 400, 110))
show_pillow_image(logo3_image)

logo2_image = openmined_image.crop((0, 110, 400, 190))
show_pillow_image(logo2_image)
```

```python
height1 = above_image.size[1]
height2 = below_image.size[1]
logos_height = 130

brain_image_size = brain_images[0].size
print('brain_image_size={}'.format(brain_image_size))
size = (brain_image_size[0] * 2, brain_image_size[1] * 3 + height1 + logos_height + height2)
design = PIL.Image.new('RGB', size, color = 0xffffff)
print('design.size={}'.format(design.size))

index = 0
for j in range(3):
    for i in range(2):
        brain_image = brain_images[index]
        index += 1
        design.paste(brain_image, (i * brain_image_size[0], j * brain_image_size[1] + height1))
        
design.paste(above_image, (0, 0))
design.paste(below_image, (0, size[1] - height2))

logo_row = brain_image_size[1] * 3 + height1
design.paste(pytorch_image, (50, logo_row))
design.paste(logo2_image, (450, logo_row + 30))
design.paste(logo3_image, (850, logo_row + 10))

show_pillow_image(design)
design.save('final_designs/DESIGN.png')
```

```python
import matplotlib.pyplot as plt
import PIL
from PIL import Image
import numpy as np

def show_pillow_image(image):
    #plt.figure() 
    plt.figure(figsize=(10, 10))
    plt.imshow(image)
    #plt.ioff()
    plt.show()

def open_image(file):
    return Image.open(file)        
        
brain_images = []
for i in range(6):
    file = 'brain_images/output-brain-{}.png'.format(i)
    brain_image = open_image(file)
    brain_images.append(brain_image)
    show_pillow_image(brain_image)
     
```

```python
laptop_file = 'design2/laptop2.png'
laptop_image = open_image(laptop_file)
laptop_image = laptop_image.crop((50, 150, 950, 850)) #.convert('RGBA')
new_size = (laptop_image.size[0] // 3, laptop_image.size[1] // 3)
laptop_image = laptop_image.resize(new_size, Image.ANTIALIAS)
print('laptop_image.size={}'.format(laptop_image.size))
laptop_width, laptop_height = laptop_image.size
show_pillow_image(laptop_image)

brain_image = brain_images[4].copy()
#brain_image.paste(laptop_image, (0, 0))
brain_width, brain_height = brain_image.size
show_pillow_image(brain_image)
```

```python
from PIL import Image, ImageOps

circuit_file = 'design2/vector-circuit.png'
circuit_image = open_image(circuit_file)
#circuit_image = ImageOps.invert(circuit_image)
#circuit_image = circuit_image.convert('LA')
circuit_image = circuit_image.crop((0, 0, 1000, 1000)) #.convert('RGBA')
new_size = (500, 500) #(250, 250) #(500, 500)
circuit_image = circuit_image.resize(new_size, Image.ANTIALIAS)
print('circuit_image.size={}'.format(circuit_image.size))
show_pillow_image(circuit_image)

alpha = 0xff000000 #0xff000000 #0xe0000000 #0xc0000000
color = 0x0b126c #6c120b #0x361620 #201636 #0x331526 #261533 #0x3e1516 #0x561911 #0x741d04 #0x111956 #0x0016153e #0x002f1428 #0x00742a00
background = PIL.Image.new('RGBA', circuit_image.size, color = alpha + color) #0xc0ffff00) # 0xc0ff0000) # 0x80ff0000)
show_pillow_image(background)

alpha = 0xe0000000 #0xe0000000 #0xc0000000
#color = 0x741d04 #0x111956 #0x0016153e #0x002f1428 #0x00742a00
background_to_mix = PIL.Image.new('RGBA', circuit_image.size, color = alpha + color)
circuit_image.paste(background_to_mix, (0, 0), mask = circuit_image)
show_pillow_image(circuit_image)
circuit_image.save('design2/bluer-vector-circuit.png')


#new_background = Image.alpha_composite(circuit_image, background)
background.paste(circuit_image, (0, 0), mask = circuit_image)
new_background = background.copy()
show_pillow_image(new_background)

N = 3
tile = PIL.Image.new('RGBA', (circuit_image.size[0] * N, circuit_image.size[1] * N))
for x in range(N):
    for y in range(N):
        tile.paste(new_background, (circuit_image.size[0] * x, circuit_image.size[1] * y))
show_pillow_image(tile)
```

```python
files = ['zero', 'one', 'eye', 'left', 'right', 'up', 'down']

icons = []
for file in files:
    image = open_image('design2/fire-icons/{}.png'.format(file))
    image = image.resize((32, 32), Image.ANTIALIAS)
    icons.append(image)
    show_pillow_image(image)
```

```python
height1 = above_image.size[1]
height2 = below_image.size[1]

brain_image_size = brain_images[0].size
design_size = (design_width, height1 + height2 + 2 * laptop_height + brain_height)
design = PIL.Image.new('RGBA', design_size, color=0xff000000)
print('design.size={}'.format(design.size))

N = 3
for x in range(N):
    for y in range(N):
        design.paste(new_background, (circuit_image.size[0] * x, circuit_image.size[1] * y))

def draw_icon_row(p, N, direction):
    previous_index = -1
    for i in range(N):
        while True:
            index = np.random.randint(0, 4)
            if index < 2 or index != previous_index: break
        previous_index = index
        index = 3 + direction if index == 3 else index
        image = icons[index]
        design.paste(image, (p[0] + i * 32, p[1]), mask = image)

def draw_icon_rows(p, N, number):
    x, y = p
    for i in range(number):
        draw_icon_row((x, y + i * 32), N, 1 if i % 2 == 0 else 0)
        
draw_icon_rows((laptop_width - 32 * 2, height1), 24, 7)
draw_icon_rows((laptop_width - 32 * 2, height1 + brain_image_size[1] + laptop_height), 24, 7)


def draw_icon_column(p, N, direction):
    previous_index = -1
    for i in range(N):
        while True:
            index = np.random.randint(0, 4)
            if index < 2 or index != previous_index: break
        previous_index = index
        index = 3 + direction if index == 3 else index
        image = icons[index]
        design.paste(image, (p[0], p[1] + i * 32), mask = image)

def draw_icon_columns(p, N, number):
    x, y = p
    for i in range(number):
        draw_icon_column((x + i * 32, y), N, 2 if i % 2 == 0 else 3)

draw_icon_columns((0, height1 + laptop_height - 32), 18, 9)
draw_icon_columns((laptop_width + brain_image_size[0], height1 + laptop_height - 32), 18, 9)
        
for x in range(2):
    for y in range(2):
        position = (x * (brain_image_size[0] + laptop_width), height1 + y * (brain_image_size[1] + laptop_height))
        design.paste(laptop_image, position, mask = laptop_image)
        
design.paste(brain_image, (laptop_width, laptop_height + height1))
design.paste(above_image, (0, 0))
design.paste(below_image, (0, design_size[1] - height2))

show_pillow_image(design)
design.save('final_designs/DESIGN4.png')
```
